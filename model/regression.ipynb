```python
#Understanding Error Surface
#All imports
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import collections

########################################
#Generate meaningfull data
########################################
def fun_Q1(x,noise):
  return 3.86*x-110.42+noise

noise=np.random.normal(0,np.sqrt(20),25)

x_q1=np.linspace(30,100,25)
y_q1=fun_Q1(x_q1,noise)


########################################
#Plot scatter plot of data
########################################
plt.scatter(x_q1,y_q1)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()



########################################
#Weight estimation through error surface, i.e., empirically locate the minima of error surface
########################################
#Sample a bunch of w's around w_opt and compute the associated error
a=3.86
b=-110.42

w0=np.linspace(-100+b,100+b,400)
w1=np.linspace(-5+a,5+a,400)

w0,w1=np.meshgrid(w0,w1)

#Compute the error
def Error(w,t,x): #inputs : 1)weight 2)data i.e (t,x)
    #Estimate the target
    error=[]
    for i in range(len(w[0])):
      p=[]
      for j in range(len(w[0])):
        e=0
        for k in range(len(x)):
          e=e+(t[k]-w[0][i][j]-w[1][i][j]*x[k])**2
        e=e/2
        p.append(e)
      error.append(p)

    error=np.array(error)

    #Compute and return the error
    return error


#Plot 3D error surface and the corresponding contour plots
#Error surface is a function of w0 and w1
E=Error([w0,w1],y_q1,x_q1)

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(w0,w1,E,alpha=0.7)


ax.set_xlabel("w0")
ax.set_ylabel("w1")
ax.set_zlabel("J(w0,w1)",fontsize=10)

ax.set_title("Error with different value of w0,w1")

#Locate the minima of the error surface
min_ind=np.unravel_index(np.argmin(E),E.shape)

min_val=E[min_ind]

min_w0=w0[min_ind]
min_w1=w1[min_ind]

# Add scatter point for minimum error
ax.scatter(min_w0, min_w1, min_val, color='red', label='Minimum Error')


plt.legend()
plt.show()


########################################
#Least squares approach to estimate the weights
########################################
#Complete the below linear regression function
def LinearRegression(x,t): #inputs : 1)input data i.e (x). 2)target i.e (t)
    X=[[1,i] for i in x]
    X=np.array(X)

    X_T=np.transpose(X)

    f=np.dot(X_T,X)
    f=np.linalg.inv(f)

    g=np.dot(X_T,t)

    w_opt=np.dot(f,g)


    return w_opt


#Estimate optimal weight's using "LinearRegression" function
w_opt=LinearRegression(x_q1,y_q1)


#Estimate the targets using the input x and the estimated weights
target=[w_opt[0] + w_opt[1]*i for i in x_q1]
target=np.array(target)


#Plot the estimated line on top of the above scatter plot
plt.scatter(x_q1,y_q1)
plt.plot(x_q1,target,color="red")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()



########################################
#Compare the estimated weight's using least squares approach with the error surface approach
########################################

print("Error surface w0 :", min_w0)
print("Error surface w1 :",min_w1)
print()
print("Least square w0 :",w_opt[0])
print("Least square w1 :",w_opt[1])



```

```python
#Understanding model order and overfitting
########################################
#Generate 20 data points
########################################
def fun_Q2(x,noise):
  return np.sin(2*np.pi*x)+ noise

noise=np.random.normal(0,np.sqrt(0.1),20)

x_q2=np.linspace(0,1,20)

y_q2=fun_Q2(x_q2,noise)


#Obtain train and test splits
X_train_q2=[x_q2[i] for i in range(0,len(x_q2),2)]
Y_train_q2=[y_q2[i] for i in range(0,len(x_q2),2)]

X_test_q2=[x_q2[i] for i in range(1,len(x_q2),2)]
Y_test_q2=[y_q2[i] for i in range(1,len(x_q2),2)]



########################################
#Fitting Mth degree polynomial using least squares approach
########################################
#Complete the function
def PolynomialFit(X_train,Y_train,M,lamda): #(training data, trining targets, Model order, Regularization coefficient)
    #Transform the data using polynomial kernel
    Q=[]
    for i in X_train:
      input=[]
      for j in range(M+1):
        c=np.power(i,j)
        input.append(c)
      Q.append(input)

    Q=np.array(Q)

    #Find Pseudo inverse solution
    Q_T=np.transpose(Q)
    I=np.eye(len(Q_T))

    f=np.dot(Q_T,Q)
    f=f+lamda*I
    f=np.linalg.inv(f)

    g=np.dot(Q_T,Y_train)

    w_opt=np.dot(f,g)

    #return the weight vector
    return w_opt

#Complete the function
def PolynomialPred(w_est,X_train,X_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    Q = []
    for i in X_train:
        input = []
        for j in range(len(w_est)):
            c = np.power(i, j)
            input.append(c)
        Q.append(input)

    Q1 = []
    for i in X_test:
        input = []
        for j in range(len(w_est)):
            c = np.power(i, j)
            input.append(c)
        Q1.append(input)

    Q = np.array(Q)
    Q1 = np.array(Q1)

    Trainpred=np.dot(Q,w_est)
    Testpred=np.dot(Q1,w_est)

    #Return training and testing predictions
    return Trainpred,Testpred

#Complete the function
def PolynomialPred_Error(w_est,X_train,X_test,Y_train,Y_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    Q = []
    for i in X_train:
        input = []
        for j in range(len(w_est)):
            c = np.power(i, j)
            input.append(c)
        Q.append(input)

    Q1 = []
    for i in X_test:
        input = []
        for j in range(len(w_est)):
            c = np.power(i, j)
            input.append(c)
        Q1.append(input)

    Q = np.array(Q)
    Q1 = np.array(Q1)

    f=np.dot(Q,w_est)
    g=np.dot(Q1,w_est)

    #Return training and testing error
    err1=Y_train-f
    TrainError=0
    for i in err1:
        TrainError=TrainError+(i*i)/2

    err2=Y_test-g
    TestError=0
    for i in err2:
        TestError=TestError+(i*i)/2

    return TrainError,TestError

#Iterate through range of M values
M_range=list(range(10))
TrError = []
TeError = []
Ta=[]
Tb=[]
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt=PolynomialFit(X_train_q2,Y_train_q2,M,lamda=0)

    #Predict training and testing targets
    Trainpred,Testpred=PolynomialPred(w_opt,X_train_q2,X_test_q2)


    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    TrainError,TestError=PolynomialPred_Error(w_opt,X_train_q2,X_test_q2,Y_train_q2,Y_test_q2)

    #Store them for plotting
    TrError.append(TrainError)
    TeError.append(TestError)

    Ta.append(Trainpred)
    Tb.append(Testpred)


#Plot training and testing estimates alogwith the original targets
col=["blue","orange","green","red","purple","brown","pink","gray","olive","cyan"]
for i in M_range:
  plt.plot(X_train_q2,Ta[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_q2,Y_train_q2,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets N=10")
plt.show()

for i in M_range:
  plt.plot(X_test_q2,Tb[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_q2,Y_test_q2,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets N=10")
plt.ylim(-1.5,2)
plt.show()



#Plot training error vs polynomial order, and testing error vs polynomial order
plt.plot(M_range,TrError,color="red",label="Training Error")
plt.plot(M_range,TeError,color="blue",label="Testing Error")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs polynomial order N=10")
plt.legend()
plt.show()



########################################
#Increase the size of training data set to 100 points and repeat the experiments
########################################
noise2=np.random.normal(0,np.sqrt(0.1),200)

x_q22=np.linspace(0,1,200)
y_q22=fun_Q2(x_q22,noise2)

X_train_q22=[x_q22[i] for i in range(0,len(x_q22),2)]
Y_train_q22=[y_q22[i] for i in range(0,len(x_q22),2)]

X_test_q22=[x_q22[i] for i in range(1,len(x_q22),2)]
Y_test_q22=[y_q22[i] for i in range(1,len(x_q22),2)]

M_range=list(range(10))
TrError = []
TeError = []
Ta=[]
Tb=[]
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt=PolynomialFit(X_train_q22,Y_train_q22,M,lamda=0)

    #Predict training and testing targets
    Trainpred,Testpred=PolynomialPred(w_opt,X_train_q22,X_test_q22)

    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    TrainError,TestError=PolynomialPred_Error(w_opt,X_train_q22,X_test_q22,Y_train_q22,Y_test_q22)

    #Store them for plotting
    TrError.append(TrainError)
    TeError.append(TestError)

    Ta.append(Trainpred)
    Tb.append(Testpred)

#Plot training and testing estimates alogwith the original targets
col=["blue","orange","green","red","purple","brown","pink","gray","olive","cyan"]
for i in M_range:
  plt.plot(X_train_q22,Ta[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_q22,Y_train_q22,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets N=100")
plt.show()

for i in M_range:
  plt.plot(X_test_q22,Tb[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_q22,Y_test_q22,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets N=100")
plt.show()



#Plot training error vs polynomial order, and testing error vs polynomial order
plt.plot(M_range,TrError,color="red",label="Training Error")
plt.plot(M_range,TeError,color="blue",label="Testing Error")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs polynomial order N=100")
plt.legend()
plt.show()


########################################
#Effect of regularization
########################################
#Consider a set of lambda's. For example: lamdas = [0, 1e-7 , 1e-4, 1e-2, 1]
#Repeat the experiments, i.e., plot the prediction and error in predictions with respect to model order. Contrast these results with those obtained without regularization.
lamdas=[0,np.exp(-7),np.exp(-4),np.exp(-2),1]

M_range=list(range(10))
TrError10 = []
TeError10 = []
TrError100 = []
TeError100 = []

Ta10=[]
Tb10=[]
Ta100=[]
Tb100=[]

for l in lamdas:
    Tr1=[]
    Tr2=[]
    Te1=[]
    Te2=[]

    Trp1=[]
    Tep1=[]
    Trp2=[]
    Tep2=[]

    for M in M_range:
        #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
        w_opt1=PolynomialFit(X_train_q2,Y_train_q2,M,l)
        w_opt2=PolynomialFit(X_train_q22,Y_train_q22,M,l)

        #Predict training and testing targets
        Trainpred1,Testpred1=PolynomialPred(w_opt1,X_train_q2,X_test_q2)
        Trainpred2,Testpred2=PolynomialPred(w_opt2,X_train_q22,X_test_q22)

        #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
        TrainError1,TestError1=PolynomialPred_Error(w_opt1,X_train_q2,X_test_q2,Y_train_q2,Y_test_q2)
        TrainError2,TestError2=PolynomialPred_Error(w_opt2,X_train_q22,X_test_q22,Y_train_q22,Y_test_q22)

        #Store them for plotting
        Tr1.append(TrainError1)
        Te1.append(TestError1)
        Tr2.append(TrainError2)
        Te2.append(TestError2)

        Trp1.append(Trainpred1)
        Tep1.append(Testpred1)
        Trp2.append(Trainpred2)
        Tep2.append(Testpred2)


    TrError10.append(Tr1)
    TeError10.append(Te1)
    TrError100.append(Tr2)
    TeError100.append(Te2)

    Ta10.append(Trp1)
    Tb10.append(Tep1)
    Ta100.append(Trp2)
    Tb100.append(Tep2)

#Plot training and testing estimates alogwith the original targets
col=["blue","orange","green","red","purple","brown","pink","gray","olive","cyan"]
for j in range(len(lamdas)):
  for i in M_range:
    plt.plot(X_train_q2,Ta10[j][i],color=col[i],label=f"Model{i}")
  plt.plot(X_train_q2,Y_train_q2,color="black",label="Original target")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(f"training and Original targets with N=10 and lamda{j+1}={lamdas[j]}")
  plt.show()


for j in range(len(lamdas)):
  for i in M_range:
    plt.plot(X_test_q2,Tb10[j][i],color=col[i],label=f"Model{i}")
  plt.plot(X_test_q2,Y_test_q2,color="black",label="Original target")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(f"testing and Original targets with N=10 and lamda{j+1}={lamdas[j]}")
  plt.ylim(-1.5,1.5)
  plt.show()

for j in range(len(lamdas)):
  for i in M_range:
    plt.plot(X_train_q22,Ta100[j][i],color=col[i],label=f"Model{i}")
  plt.plot(X_train_q22,Y_train_q22,color="black",label="Original target")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(f"training and Original targets with N=100 and lamda{j+1}={lamdas[j]}")
  plt.show()

for j in range(len(lamdas)):
  for i in M_range:
    plt.plot(X_test_q22,Tb100[j][i],color=col[i],label=f"Model{i}")
  plt.plot(X_test_q22,Y_test_q22,color="black",label="Original target")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(f"testing and Original targets with N=100 and lamda{j+1}={lamdas[j]}")
  plt.show()



#Plot training error vs polynomial order, and testing error vs polynomial order
for i in range(len(TrError10)):
  plt.plot(M_range,TrError10[i],label=f"lamdas{i+1}={lamdas[i]}")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs M Given Training N=10")
plt.legend()
plt.show()

for i in range(len(TeError10)):
  plt.plot(M_range,TeError10[i],label=f"lamdas{i}={lamdas[i]}")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs M Given Testing N=10")
plt.legend()
plt.show()

for i in range(len(TrError100)):
  plt.plot(M_range,TrError100[i],label=f"lamdas{i}={lamdas[i]}")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs M Given Training N=100")
plt.legend()
plt.show()

for i in range(len(TeError100)):
  plt.plot(M_range,TeError100[i],label=f"lamdas{i}={lamdas[i]}")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs M Given Testing N=100")
plt.legend()
plt.show()


########################################
#Effect of bias regularization
########################################
#Modify the function i.e include bias
#Generate data
def fun_Q23(x,noise):
  return 5+np.sin(2*np.pi*x)+noise

noise3=np.random.normal(0,np.sqrt(0.1),50)

x_q23=np.linspace(0,1,50)
y_q23=fun_Q23(x_q23,noise3)

X_train_q23=[x_q23[i] for i in range(0,len(x_q23),2)]
Y_train_q23=[y_q23[i] for i in range(0,len(x_q23),2)]

X_test_q23=[x_q23[i] for i in range(1,len(x_q23),2)]
Y_test_q23=[y_q23[i] for i in range(1,len(x_q23),2)]


#Estimate the polynomial with and without regularization constraint
#Compare the two estimated polynomials and report the observations
lam=[0,np.exp(-4)]

TrError= []
TeError= []
Ta=[]
Tb=[]

for l in lam:
    Tr=[]
    Te=[]

    Trp=[]
    Tep=[]

    for M in M_range:
        #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
        w_opt=PolynomialFit(X_train_q23,Y_train_q23,M,l)

        #Predict training and testing targets
        Trainpred,Testpred=PolynomialPred(w_opt,X_train_q23,X_test_q23)

        #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
        TrainError,TestError=PolynomialPred_Error(w_opt,X_train_q23,X_test_q23,Y_train_q23,Y_test_q23)

        #Store them for plotting
        Tr.append(TrainError)
        Te.append(TestError)

        Trp.append(Trainpred)
        Tep.append(Testpred)


    TrError.append(Tr)
    TeError.append(Te)

    Ta.append(Trp)
    Tb.append(Tep)

col=["blue","orange","green","red","purple","brown","pink","gray","olive","cyan"]

for j in range(len(lam)):
  for i in M_range:
    plt.plot(X_train_q23,Ta[j][i],color=col[i],label=f"Model{i}")
  plt.plot(X_train_q23,Y_train_q23,color="black",label="Original target")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(f"Bias training and Original targets with N=25 and lamda{j+1}={lamdas[j]}")
  plt.show()


for j in range(len(lam)):
  for i in M_range:
    plt.plot(X_test_q23,Tb[j][i],color=col[i],label=f"Model{i}")
  plt.plot(X_test_q23,Y_test_q23,color="black",label="Original target")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(f"Bias testing and Original targets with N=25 and lamda{j+1}={lamdas[j]}")
  plt.show()

for i in range(len(TrError)):
  plt.plot(M_range,TrError[i],label=f"lamdas{i+1}={lamdas[i]}")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Bias \Error vs M Given Training N=25")
plt.legend()
plt.show()

for i in range(len(TeError)):
  plt.plot(M_range,TeError[i],label=f"lamdas{i+1}={lamdas[i]}")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Bias \Error vs M Given Testing N=25")
plt.legend()
plt.show()

```

```python
#Understanding the choice of kernel
########################################
#Generate 100 data points
########################################
import numpy as np
import matplotlib.pyplot as plt

def fun_Q3(x,noise):
  return np.sin(2*np.pi*x)+noise

noise_3=np.random.normal(0,np.sqrt(0.1),100)

x_q3=np.linspace(0,1,100)
y_q3=fun_Q3(x_q3,noise_3)


#Obtian train and test splits
#Take even samples for training and odd samples for testing
X_train_q3=[x_q3[i] for i in range(0,len(x_q3),2)]
Y_train_q3=[y_q3[i] for i in range(0,len(x_q3),2)]

X_test_q3=[x_q3[i] for i in range(1,len(x_q3),2)]
Y_test_q3=[y_q3[i] for i in range(1,len(x_q3),2)]




#Function to estimate the parameters
def KernelRegressionFit(X_train,Y_train,kernelType,M,lamda): #(training data, training targets, type of kernel, regularization coefficient)
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    #print(X_train.shape)
    #X_train = np.reshape(X_train, (1,-1))
    #Y_train = np.reshape(Y_train, (1,-1))
    X = []
    if kernelType=='polynomial':
      #Use polynomial kernel to transform the data
      Q=[]
      for i in X_train:
        input=[]
        for j in range(M+1):
            c=np.power(i,j)
            input.append(c)
        Q.append(input)
      Q=np.array(Q)

    if kernelType=='gaussian':
      #Use Gaussian kernel to transform the data
      u=np.linspace(0,1,M)
      sigma=1
      Q=[]
      for i in X_train:
        input=[1]
        for j in range(M):
          c=np.exp(-(i-u[j])**2/(2*sigma**2))
          input.append(c)
        Q.append(input)
      Q=np.array(Q)

    if kernelType=='sigmoidal':
      #Use Sigmoid kernel to transform the data
      u=np.linspace(0,1,M)
      s=1
      Q=[]
      for i in X_train:
        input=[1]
        for j in range(M):
          c=1/(1+np.exp(-(i-u[j])/s))
          input.append(c)
        Q.append(input)
      Q=np.array(Q)

    #Estimate weights using Pseudo iverse solution
    Q_T=np.transpose(Q)

    I=np.eye(len(Q_T))

    f=np.dot(Q_T,Q)
    f=f+lamda*I
    f=np.linalg.inv(f)
    g=np.dot(Q_T,Y_train)

    w_opt=np.dot(f,g)

    #Return the estimated weights
    return w_opt

#Function to compute the training and testing errors from the current weight estimates
def KernelRegressionPred_Error(w_est,X_train,Y_train,X_test,Y_test,kernelType):
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    X_tr = []
    X_te = []
    #X_train = np.reshape(X_train, (1,-1))
    #Y_train = np.reshape(Y_train, (1,-1))
    #X_test = np.reshape(X_test, (1,-1))
    #Y_test = np.reshape(Y_test, (1,-1))
    M = len(w_est)-1
    if kernelType=='polynomial':
      #Use polynomial kernel to transform the data
      Q=[]
      for i in X_train:
        input=[]
        for j in range(M+1):
            c=np.power(i,j)
            input.append(c)
        Q.append(input)

      Q1=[]
      for i in X_test:
        input=[]
        for j in range(M+1):
            c=np.power(i,j)
            input.append(c)
        Q1.append(input)

      Q=np.array(Q)
      Q1 = np.array(Q1)

    if kernelType=='gaussian':
      #Use Gaussian kernel to transform the data
      u=np.linspace(0,1,M)
      sigma=1
      Q=[]
      for i in X_train:
        input=[1]
        for j in range(M):
          c=np.exp(-(i-u[j])**2/(2*sigma**2))
          input.append(c)
        Q.append(input)

      Q1=[]
      for i in X_test:
        input=[1]
        for j in range(M):
          c=np.exp(-(i-u[j])**2/(2*sigma**2))
          input.append(c)
        Q1.append(input)

      Q=np.array(Q)
      Q1=np.array(Q1)

    if kernelType=='sigmoidal':
      #Use Sigmoid kernel to transform the data
      u=np.linspace(0,1,M)
      s=1
      Q=[]
      for i in X_train:
        input=[1]
        for j in range(M):
          c=1/(1+np.exp(-(i-u[j])/s))
          input.append(c)
        Q.append(input)

      Q1=[]
      for i in X_test:
        input=[1]
        for j in range(M):
          c=1/(1+np.exp(-(i-u[j])/s))
          input.append(c)
        Q1.append(input)

      Q=np.array(Q)
      Q1=np.array(Q1)


    #Estimate training and testing targets
    f=np.dot(Q,w_est)
    g=np.dot(Q1,w_est)


    #Compute and return the training and testing errors
    err1=Y_train-f
    TrainError=0
    for i in err1:
      TrainError=TrainError+(i*i)/2

    err2=Y_test-g
    TestError=0
    for i in err2:
      TestError=TestError+(i*i)/2

    return TrainError,TestError

#fun to estimate targets
def KernelRegressionPred(w_est,X_train,X_test,kernelType):
  M = len(w_est)-1
  if kernelType=='polynomial':
    #Use polynomial kernel to transform the data
    Q=[]
    for i in X_train:
      input=[]
      for j in range(M+1):
          c=np.power(i,j)
          input.append(c)
      Q.append(input)

    Q1=[]
    for i in X_test:
        input=[]
        for j in range(M+1):
            c=np.power(i,j)
            input.append(c)
        Q1.append(input)

    Q=np.array(Q)
    Q1 = np.array(Q1)

  if kernelType=='gaussian':
      #Use Gaussian kernel to transform the data
      u=np.linspace(0,1,M)
      sigma=1
      Q=[]
      for i in X_train:
        input=[1]
        for j in range(M):
          c=np.exp(-(i-u[j])**2/(2*sigma**2))
          input.append(c)
        Q.append(input)

      Q1=[]
      for i in X_test:
        input=[1]
        for j in range(M):
          c=np.exp(-(i-u[j])**2/(2*sigma**2))
          input.append(c)
        Q1.append(input)

      Q=np.array(Q)
      Q1=np.array(Q1)

  if kernelType=='sigmoidal':
      #Use Sigmoid kernel to transform the data
      u=np.linspace(0,1,M)
      s=1
      Q=[]
      for i in X_train:
        input=[1]
        for j in range(M):
          c=1/(1+np.exp(-(i-u[j])/s))
          input.append(c)
        Q.append(input)

      Q1=[]
      for i in X_test:
        input=[1]
        for j in range(M):
          c=1/(1+np.exp(-(i-u[j])/s))
          input.append(c)
        Q1.append(input)

      Q=np.array(Q)
      Q1=np.array(Q1)

  #Estimate training and testing targets
  Trainpred=np.dot(Q,w_est)
  Testpred=np.dot(Q1,w_est)

  return Trainpred,Testpred


#Iterate through range of M values
M_range=list(range(10))

polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoid_tr_error = []
sigmoid_te_error = []

Trp1=[]
Tep1=[]
Trp2=[]
Tep2=[]
Trp3=[]
Tep3=[]

for M in M_range:
    #Fit Mth order polynomial using three kernels i.e {Polynomial,Gaussian,Sigmoid}
    w_opt1=KernelRegressionFit(X_train_q3,Y_train_q3,kernelType='polynomial',M=M,lamda=0)
    w_opt2=KernelRegressionFit(X_train_q3,Y_train_q3,kernelType='gaussian',M=M,lamda=0)
    w_opt3=KernelRegressionFit(X_train_q3,Y_train_q3,kernelType='sigmoidal',M=M,lamda=0)

    #Predict training and testing targets using estimated w
    Trainpred1,Testpred1=KernelRegressionPred(w_opt1,X_train_q3,X_test_q3,kernelType='polynomial')
    Trainpred2,Testpred2=KernelRegressionPred(w_opt2,X_train_q3,X_test_q3,kernelType='gaussian')
    Trainpred3,Testpred3=KernelRegressionPred(w_opt3,X_train_q3,X_test_q3,kernelType='sigmoidal')

    #Predict errors on both training and testing data using estimated w
    TrainError1,TestError1=KernelRegressionPred_Error(w_opt1,X_train_q3,Y_train_q3,X_test_q3,Y_test_q3,kernelType='polynomial')
    TrainError2,TestError2=KernelRegressionPred_Error(w_opt2,X_train_q3,Y_train_q3,X_test_q3,Y_test_q3,kernelType='gaussian')
    TrainError3,TestError3=KernelRegressionPred_Error(w_opt3,X_train_q3,Y_train_q3,X_test_q3,Y_test_q3,kernelType='sigmoidal')

    #Store them for plotting
    polynomial_tr_error.append(TrainError1)
    polynomial_te_error.append(TestError1)
    gaussian_tr_error.append(TrainError2)
    gaussian_te_error.append(TestError2)
    sigmoid_tr_error.append(TrainError3)
    sigmoid_te_error.append(TestError3)

    Trp1.append(Trainpred1)
    Tep1.append(Testpred1)
    Trp2.append(Trainpred2)
    Tep2.append(Testpred2)
    Trp3.append(Trainpred3)
    Tep3.append(Testpred3)

#Plot the predicted training and testing targets alongside the original targets for various model orders and all three different kernels.
col=["blue","orange","green","red","purple","brown","pink","gray","olive","cyan"]
for i in M_range:
  plt.plot(X_train_q3,Trp1[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_q3,Y_train_q3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets of Polynomial N=50")
plt.show()

for i in M_range:
  plt.plot(X_test_q3,Tep1[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_q3,Y_test_q3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets of polynoimal N=50")
plt.show()

for i in M_range:
  plt.plot(X_train_q3,Trp2[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_q3,Y_train_q3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets of gaussian N=50")
plt.show()

for i in M_range:
  plt.plot(X_test_q3,Tep2[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_q3,Y_test_q3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets of gaussian N=50")
plt.show()

for i in M_range:
  plt.plot(X_train_q3,Trp3[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_q3,Y_train_q3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets of sigmoid N=50")
plt.show()

for i in M_range:
  plt.plot(X_test_q3,Tep3[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_q3,Y_test_q3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets of sigmoid N=50")
plt.show()




#Plot training error vs polynomial order and testing error vs polynomial order for all the three different kernels
plt.plot(M_range,polynomial_tr_error,color="red",label="polynomial_training")
plt.plot(M_range,polynomial_te_error,color="blue",label="polynomial_testing")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs M N=50")
plt.legend()
plt.show()

plt.plot(M_range,gaussian_tr_error,color="red",label="gaussian_training")
plt.plot(M_range,gaussian_te_error,color="blue",label="gaussian_testing")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs M N=50")
plt.legend()
plt.show()

plt.plot(M_range,sigmoid_tr_error,color="red",label="sigmoid_training")
plt.plot(M_range,sigmoid_te_error,color="blue",label="sigmoid_testing")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error vs M N=50")
plt.legend()
plt.show()



########################################
#Repeat the experiments by changing target function
########################################
def fun(x):
  if x>=0 and x<1:
    return np.sin(2*3.14*x)
  elif x>=1 and x<2:
    if x<1.5:
      return 3*(x-1)
    else:
      return -3*(x-2)
  else:
    m=2.5
    sigma=.15
    return np.exp(-(x-m)**2/(2*sigma**2))


noise_3=np.random.normal(0,np.sqrt(0.1),100)
x_3=np.linspace(0,3,100)
y_3=[fun(i) for i in x_3]
y_3=np.array(y_3)
y_3=y_3+noise_3

X_train_3=[x_3[i] for i in range(0,len(x_3),2)]
Y_train_3=[y_3[i] for i in range(0,len(x_3),2)]
X_test_3=[x_3[i] for i in range(1,len(x_3),2)]
Y_test_3=[y_3[i] for i in range(1,len(x_3),2)]

M_range=list(range(10))

polynomial_tr_pred = []
polynomial_te_pred = []
gaussian_tr_pred = []
gaussian_te_pred = []
sigmoid_tr_pred = []
sigmoid_te_pred = []

polynomial_tr_error=[]
polynomial_te_error=[]
gaussian_tr_error=[]
gaussian_te_error=[]
sigmoid_tr_error=[]
sigmoid_te_error=[]


for i in M_range:
  w_opt1=KernelRegressionFit(X_train_3,Y_train_3,kernelType='polynomial',M=i,lamda=0)
  Trainpred1,Testpred1=KernelRegressionPred(w_opt1,X_train_3,X_test_3,kernelType='polynomial')
  TrainError1,TestError1=KernelRegressionPred_Error(w_opt1,X_train_3,Y_train_3,X_test_3,Y_test_3,kernelType='polynomial')

  w_opt2=KernelRegressionFit(X_train_3,Y_train_3,kernelType='gaussian',M=i,lamda=0)
  Trainpred2,Testpred2=KernelRegressionPred(w_opt2,X_train_3,X_test_3,kernelType='gaussian')
  TrainError2,TestError2=KernelRegressionPred_Error(w_opt2,X_train_3,Y_train_3,X_test_3,Y_test_3,kernelType='gaussian')

  w_opt3=KernelRegressionFit(X_train_3,Y_train_3,kernelType='sigmoidal',M=i,lamda=0)
  Trainpred3,Testpred3=KernelRegressionPred(w_opt3,X_train_3,X_test_3,kernelType='sigmoidal')
  TrainError3,TestError3=KernelRegressionPred_Error(w_opt3,X_train_3,Y_train_3,X_test_3,Y_test_3,kernelType='sigmoidal')

  polynomial_tr_error.append(TrainError1)
  polynomial_te_error.append(TestError1)
  gaussian_tr_error.append(TrainError2)
  gaussian_te_error.append(TestError2)
  sigmoid_tr_error.append(TrainError3)
  sigmoid_te_error.append(TestError3)

  polynomial_tr_pred.append(Trainpred1)
  polynomial_te_pred.append(Testpred1)
  gaussian_tr_pred.append(Trainpred2)
  gaussian_te_pred.append(Testpred2)
  sigmoid_tr_pred.append(Trainpred3)
  sigmoid_te_pred.append(Testpred3)


#Plot the predicted training and testing targets alongside the original targets for various model orders and all three different kernels.
col=["blue","orange","green","red","purple","brown","pink","gray","olive","cyan"]

for i in M_range:
  plt.plot(X_train_3,polynomial_tr_pred[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_3,Y_train_3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets of Polynomial N=50")
plt.show()

for i in M_range:
  plt.plot(X_test_3,polynomial_te_pred[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_3,Y_test_3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets of polynoimal N=50")
plt.show()

for i in M_range:
  plt.plot(X_train_3,gaussian_tr_pred[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_3,Y_train_3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets of gaussian N=50")
plt.show()

for i in M_range:
  plt.plot(X_test_3,gaussian_te_pred[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_3,Y_test_3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets of gaussian N=50")
plt.show()

for i in M_range:
  plt.plot(X_train_3,sigmoid_tr_pred[i],color=col[i],label=f"Model{i}")
plt.plot(X_train_3,Y_train_3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("training est and Original targets of sigmoid N=50")
plt.show()

for i in M_range:
  plt.plot(X_test_3,sigmoid_te_pred[i],color=col[i],label=f"Model{i}")
plt.plot(X_test_3,Y_test_3,color="black",label="Original target")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("testing est and Original targets of sigmoid N=50")
plt.show()

#Plot training error vs polynomial order and testing error vs polynomial order for all the three different kernels
plt.plot(M_range,polynomial_tr_error,color="red",label="Polynomial_training")
plt.plot(M_range,polynomial_te_error,color="blue",label="polynoial_testing")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error of funtion on polynomial kernel N=50")
plt.legend()
plt.show()

plt.plot(M_range,gaussian_tr_error,color="red",label="gaussian_training")
plt.plot(M_range,gaussian_te_error,color="blue",label="gaussian_testing")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error of funtion on gaussian kernel N=50")
plt.legend()
plt.show()

plt.plot(M_range,sigmoid_tr_error,color="red",label="sigmoid_training")
plt.plot(M_range,sigmoid_te_error,color="blue",label="sigmoid_testing")
plt.xlabel("Value of M")
plt.ylabel("Error")
plt.title("Error of funtion on sigmoid kernel N=50")
plt.legend()
plt.show()

```

```python
#Understanding the bias-variance trade-off
########################################
#Generate 100 data sets of noisy sinusoidal data
########################################
def fun_Q5(x,noise):
  return np.sin(2*np.pi*x)+noise

l_x=[]
l_y=[]
for i in range(100):
  noise5=np.random.normal(0,np.sqrt(0.1),25)
  x_q5=np.linspace(0,1,25)
  y_q5=fun_Q5(x_q5,noise5)

  l_x.append(x_q5)
  l_y.append(y_q5)


########################################
#Use regularized least squares to estimate w
########################################
def ls(x_train,y_train,M,lamda):
  u=np.linspace(0,1,M)
  sigma=1
  Q=[]
  for i in x_train:
    input=[1]
    for j in range(M):
      c=np.exp(-(i-u[j])**2/(2*sigma**2))
      input.append(c)
    Q.append(input)
  Q=np.array(Q)

  Q_T=np.transpose(Q)

  I=np.eye(len(Q_T))

  f=np.dot(Q_T,Q)
  f=f+lamda*I
  f=np.linalg.inv(f)
  g=np.dot(Q_T,y_train)

  w_opt=np.dot(f,g)

  #Return the estimated weights
  return w_opt

def targetpred(w_est,x_train,M,lamda):
  u=np.linspace(0,1,M)
  sigma=1
  Q=[]
  for i in x_train:
    input=[1]
    for j in range(M):
      c=np.exp(-(i-u[j])**2/(2*sigma**2))
      input.append(c)
    Q.append(input)
  Q=np.array(Q)

  pred=np.dot(Q,w_est)

  return pred



########################################
#Illustrate the concept of Bias-Variance trade off
########################################
#1. Chose three different regularization coefficeints (low,middle and high)
#2. For every regularization coefficient, produce two plots: one displaying 100 estimated curves, and the other showing the mean of the estimated curves alongside the original function.
#3. For three regularization coefficients, you should have a total of six plots, meaning two plots for each regularization.
#4. Using the six plots above, describe the bias-variance trade-off.
lam=[np.exp(-18),np.exp(-5),1]

k=[]
for i in range(len(lam)):
  p=[]
  for j in range(100):
    w_opt=ls(l_x[j],l_y[j],M=24,lamda=lam[i])
    pred=targetpred(w_opt,l_x[i],M=24,lamda=lam[i])
    p.append(pred)
  k.append(p)

for i in range(100):
  plt.plot(l_x[i],k[0][i])
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Lambda={lam[0]}")
plt.ylim(-4,4)
plt.show()

pred_mean = np.mean(k[0], axis=0)
plt.plot(l_x[0], pred_mean, label='Mean of estimated curves')
plt.plot(l_x[0], fun_Q5(l_x[0], 0), label='Original function', linestyle='--')
plt.legend()
plt.title(f'Mean of estimated curves vs. Original function ( lambda={lam[0]})')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,2)
plt.show()

for i in range(100):
  plt.plot(l_x[i],k[1][i])
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Lambda={lam[1]}")
plt.ylim(-2,2)
plt.show()

pred_mean = np.mean(k[1], axis=0)
plt.plot(l_x[0], pred_mean, label='Mean of estimated curves')
plt.plot(l_x[0], fun_Q5(l_x[0], 0), label='Original function', linestyle='--')
plt.legend()
plt.title(f'Mean of estimated curves vs. Original function ( lambda={lam[1]})')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5,1.5)
plt.show()

for i in range(100):
  plt.plot(l_x[i],k[2][i])
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Lambda={lam[2]}")
plt.ylim(-1.5,1.5)
plt.show()

pred_mean = np.mean(k[2], axis=0)
plt.plot(l_x[0], pred_mean, label='Mean of estimated curves')
plt.plot(l_x[0], fun_Q5(l_x[0], 0), label='Original function', linestyle='--')
plt.legend()
plt.title(f'Mean of estimated curves vs. Original function ( lambda={lam[2]})')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5,1.5)
plt.show()


```

```python
#Understanding MAP estimate
########################################
#Generate 100 data sets of noisy sinusoidal data
########################################
import numpy as np
import matplotlib.pyplot as plt

def fun_Q6(x):
  return np.sin(2*np.pi*x)

noise_6=np.random.normal(0,np.sqrt(0.1),100)

x_q6=np.linspace(0,1,100)

y_q6=fun_Q6(x_q6)
y_q6=y_q6+noise_6

#kernel
M=20
u=np.linspace(0,1,M + 1)
sigma=1

kernel=[]
for i in x_q6:
  input=[]
  for j in range(M + 1):
    c=np.exp(-(i-u[j])**2/(2*sigma**2))
    input.append(c)
  kernel.append(input)

kernel=np.array(kernel)

########################################
#Update the statistics of posterior density
########################################
#Initialie the parameters for standard normal prior
alpha = 1
mean=np.zeros(21)
covariance=(1 / alpha)*np.eye(21)

#Iterate through the data points and update the stats of posterior density
M=20
u=np.linspace(0,1,M + 1)
sigma=1

for i in range(len(x_q6)):
  Q=[]
  for j in range(M+1):
    c=np.exp(-(x_q6[i]-u[j])**2/(2*sigma**2))
    Q.append(c)

  Q=np.array(Q)
  Q_T=np.transpose(Q)

  B = 1000000
  old_cov=covariance
  #new covariance
  covariance = np.linalg.inv(np.linalg.inv(covariance) + B*np.outer(Q,Q_T))

  #new mean
  a=np.dot(np.linalg.inv(old_cov),mean)+B*np.dot(Q,y_q6[i])
  mean=np.dot(covariance,a)

########################################
#Sample weight vector from posterior distribution. Estimate the curve, repeat the procedure for 100 times and get the avg fit
########################################
sample=100
curve=[]

for i in range(sample):
  w_sample = np.random.multivariate_normal(mean,covariance)
  y_est=np.dot(kernel,w_sample)
  curve.append(y_est)

y_est_avg=np.mean(curve,axis=0)


plt.plot(x_q6,y_est_avg, color="red",label='Average Fit')
plt.scatter(x_q6, y_q6, label='Data')
plt.legend()
plt.show()


########################################
#Predictive distribution analysis
########################################
#Predictive distribution analysis through sampling
#Iterate through data points and sample weight vectors when partial data points are seen, and plot the curves
# Plot the predictive distribution through sampling
x_pred=np.linspace(0,1,5)

M=20
u_pred=np.linspace(0,1,M + 1)
sigma=1

for i in range(len(x_pred)):
  Q=[]
  for j in range(M+1):
    c=np.exp(-(x_pred[i]-u_pred[j])**2/(2*sigma**2))
    Q.append(c)

  Q=np.array(Q)

  new_mean=np.dot(np.transpose(mean),Q)

  new_variance=(1/B)+np.dot(np.dot(np.transpose(Q),covariance),Q)
  sigma=np.sqrt(new_variance)

  x = np.linspace(new_mean - 3*sigma, new_mean + 3*sigma, 1000)
  y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - new_mean)**2 / (2 * new_variance))

  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('Probability Density')
  plt.title('Normal Distribution')
  plt.grid(True)
  plt.show()


#Predictive distribution analysis through variance
#Iterate through data points and obtain necessary plots as discussed in the class




```

```python
#Real world problems
#Part(2): Breast Cancer Wisconsin (Prognostic) Data Set

from google.colab import drive
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

drive.mount('/content/drive',force_remount=True)
#Download the data from the given link and keep it in the following location
datapath = "/content/drive/MyDrive/PRML_Assignments/Assignment_2/BreastCancer/wpbc.data"
data = []
with open(datapath) as fp:
    Lines = fp.readlines()
    for line in Lines:
      data.append(line.strip().split(',')[2:33])
data = np.array(data)
data = data.astype('float32')
print(data.shape)

#Train and test splits
X=data[:,1:]
Y=data[:,0]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)



def Leastsquareregression(x,t,lamda): #inputs : 1)input data i.e (x). 2)target i.e (t)
    X=[]
    for i in range(len(x)):
      input=[1]
      for j in range(len(x[0])):
        input.append(x[i][j])
      X.append(input)

    X=np.array(X)

    X_T=np.transpose(X)
    I=np.eye(len(X_T))

    f=np.dot(X_T,X)
    f=f+lamda*I
    f=np.linalg.inv(f)

    g=np.dot(X_T,t)

    w_opt=np.dot(f,g)
    return w_opt

def Leastsquareregression_error(w_est,X_train,X_test,Y_train,Y_test):
  X1=[]
  for i in range(len(X_train)):
    input=[1]
    for j in range(len(X_train[0])):
      input.append(X_train[i][j])
    X1.append(input)

  X1=np.array(X1)

  X2=[]
  for i in range(len(X_test)):
    input=[1]
    for j in range(len(X_test[0])):
      input.append(X_test[i][j])
    X2.append(input)

  X2=np.array(X2)

  f=np.dot(X1,w_est)
  g=np.dot(X2,w_est)

  #Return training and testing error
  TrainError = np.mean((Y_train - f) ** 2)

  TestError = np.mean((Y_test - g) ** 2)

  return TrainError,TestError


lamb=[0,np.exp(-7),np.exp(-4),np.exp(-2),1]
TrError = []
TeError = []
for l in lamb:
    w_opt=Leastsquareregression(X_train,Y_train,lamda=l)

    TrainError,TestError=Leastsquareregression_error(w_opt,X_train,X_test,Y_train,Y_test)

    #Store them for plotting
    TrError.append(TrainError)
    TeError.append(TestError)




#Plot training error vs polynomial order, and testing error vs polynomial order
plt.plot(lamb,TrError,color="red",label="Training Error")
plt.plot(lamb,TeError,color="blue",label="Testing Error")
plt.xlabel("Value of lambda")
plt.ylabel("Error")
plt.title("Error vs lambda ")
plt.legend()
plt.show()






```
