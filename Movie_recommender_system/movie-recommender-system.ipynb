```python
train1=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
train2=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
train1=train1.merge(train2,on='title')
train1.shape
```

```python
train1=train1[['movie_id','title','genres','keywords','cast','crew']]
```

```python
train1.head()
```

```python
train1.isnull().sum()
```

```python
train1.duplicated().sum()
```

```python
import ast #for converting str to list

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
train1['genres'] = train1['genres'].apply(convert)
train1['keywords'] = train1['keywords'].apply(convert)
```

```python
train1.head()
```

```python
# Here i am just keeping top 3 cast

def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
train1['cast'] = train1['cast'].apply(convert_cast)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
train1['crew'] = train1['crew'].apply(fetch_director)
train1.head()
```

```python
train1['title1'] = train1['title'].apply(lambda x:x.split())
```

```python
# now removing space like that 
'Anna Kendrick'
'AnnaKendrick'

def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    
    L2=[]
    for i in L1:
        L2.append(i.lower())
    return L2
```

```python
train1['cast'] = train1['cast'].apply(remove_space)
train1['crew'] = train1['crew'].apply(remove_space)
train1['genres'] = train1['genres'].apply(remove_space)
train1['keywords'] = train1['keywords'].apply(remove_space)
train1['title1'] = train1['title1'].apply(remove_space)
```

```python
train1.head()
```

```python
train1['tag']=train1['genres']+train1['keywords']+train1['cast']+train1['crew']+train1['title1']
```

```python
train1=train1[['movie_id','title','tag']]
```

```python
train1.head(2)
```

```python
train1['tag'] = train1['tag'].apply(lambda x: " ".join(x))
train1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
vector = cv.fit_transform(train1['tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    if movie not in train1['title'].values:
        for i in range(5):
            x=np.random.randint(0, len(train1))
            print(train1.iloc[x]['title'])
    else:
        index = train1[train1['title'] == movie].index[0]
        distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
        for i in distances[1:6]:
            print(train1.iloc[i[0]].title)
```

```python
recommend("Pirates of the Caribbean: At World's End")
```

```python
recommend("John Carter")
```
